```python
Adj_close_price = google_data[["Adj Close"]]
```

```python
max(Adj_close_price.values), min(Adj_close_price.values)
```

### Now we need to convert these values between 0 and 1, called normalization.
### we got a max value of 198.160 and a min value of 4.342 we will convert all the values between 0 and 1 for training of the data faster. Because big values takes more time to train and requires more computational power. Becuase of that we normalize.
### to do that we need to import the min max scaler from sklearn

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
scaled_data = scaler.fit_transform(Adj_close_price)
scaled_data
```

```python
len(scaled_data)
```

```python
x_data = []
y_data = []
for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)
```

### here we are considering moving average for 100 days. so range starts from 100 to len(scaled_data) as the first 99 values are null values.
### here we will train the 0 to 100 values to predict the 101th value, similarly 2 to 101 values to predict 102th value, it goes till the length of the scaled data i.e 5033.


```python
x_data[0],y_data[0]
```

## Split the data into training and testing dataset

```python
int(len(x_data)*0.7) #traindata
```

```python
5033 - 100 - int(len(x_data)*0.7) #remove 100 as it has null values(MA) from the total data of 5033.It is testdata
```

```python
splitting_len = int(len(x_data)*0.7)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]

x_test = x_data[splitting_len:]
y_test = y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Building Neural Network Model

```python
#!pip install keras
```

```python
#!pip install tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()
model.add(LSTM(128, return_sequences = True, input_shape = (x_train.shape[1],1)))
model.add(LSTM(64, return_sequences = False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer = "adam", loss="mean_squared_error")
```

```python
model.fit(x_train, y_train, batch_size = 1, epochs = 2)
```

```python
model.summary()
```

## Predicting the test data

```python
predictions = model.predict(x_test)
```

```python
predictions
```

**Now we need to inverse transfrom them to get the original stock data**

```python
inv_predictions = scaler.inverse_transform(predictions)
inv_predictions
```

```python
y_test
```

**similarly we need to inverse transform them**

```python
inv_y_test = scaler.inverse_transform(y_test)
inv_y_test
```

```python
rmse = np.sqrt(np.mean((inv_predictions - inv_y_test)**2))
```

```python
rmse
```

```python
ploting_data = pd.DataFrame(
    {
        "orginial_test_data": inv_y_test.reshape(-1), #as inv_y_test is 2D array. we need to convert it to 1D array
        "predicted_test_data": inv_predictions.reshape(-1)
    },
    index = range(len(inv_y_test))
)
ploting_data.head()
```
